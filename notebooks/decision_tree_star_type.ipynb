Import Libraries

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

Dataset

```python
data = pd.read_csv('6 class csv.csv')
print(data.head())
```

```python
X = data.drop("Star type", axis=1)  #features
y = data["Star type"]   #target
print(X.head())
```

```python
set(X["Star color"])
```

```python
l = ["A","B","F","G","K","M","O"]
c = len(l)
print("{")
for i in range(c):
    t = [0] * c
    t[i] = 1
    print(f"    \"{l[i]}\": {t[1:]},")
print("}")
```

```python
X = pd.get_dummies(X, drop_first=True)
print(X.head)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
model = DecisionTreeClassifier(random_state=42)
```

```python
model.fit(X_train, y_train)
```

```python
y_predict = model.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_predict)
print("Accuracy: {:.2f}%".format(accuracy * 100))
```

```python
print("Classification Report")
print(classification_report(y_test, y_predict))
```

```python
import pickle

with open("decision_tree.pkl", "wb") as file:  # "wb" = write binary mode
    pickle.dump(model, file)  # Serialize the model to the file
```
